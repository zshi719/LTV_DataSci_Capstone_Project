## Creating Validation Set
- Last 10% of training data

```python
ne_train_tune = ne_train[:135]
ne_val_tune = ne_train[135:]

s_train_tune = s_train[:135]
s_val_tune = s_train[135:]
```

## Data Windowing
- 7 day sequence
- Creating the tf.dataset object

```python
def window_dataset(data, n_steps, n_horizon, batch_size, multi_var=False, expand_dims=False):
    #create a window with n steps back plus the size of the prediction length
    window = n_steps + n_horizon
    
    #expand dimensions to 3D to fit with LSTM inputs
    #creat the inital tensor dataset
    if expand_dims:
        ds = tf.expand_dims(data, axis=-1)
        ds = tf.data.Dataset.from_tensor_slices(ds)
    else:
        ds = tf.data.Dataset.from_tensor_slices(data)
    
    #create the window function shifting the data by the prediction length
    ds = ds.window(window, shift=n_horizon, drop_remainder=True)
    
    #flatten the dataset and batch into the window size
    ds = ds.flat_map(lambda x : x.batch(window)) 
    
    #create the supervised learning problem x and y and batch
    if multi_var:
        ds = ds.map(lambda x : (x[:-n_horizon], x[-n_horizon:, :1]))
    else:
        ds = ds.map(lambda x : (x[:-n_horizon], x[-n_horizon:]))
    
    ds = ds.batch(batch_size).prefetch(1)
    
    return ds
```

```python
tf.random.set_seed(42)

n_steps = 7
n_horizon = 1
batch_size = 8

# Data windowing on train data
ne_train_ds = window_dataset(ne_train_tune, n_steps, n_horizon, batch_size, multi_var=True)
s_train_ds = window_dataset(s_train_tune, n_steps, n_horizon, batch_size, multi_var=True)

# Data windowing on validation data
ne_val_ds = window_dataset(ne_val_tune, n_steps, n_horizon, batch_size, multi_var=True)
s_val_ds = window_dataset(s_val_tune, n_steps, n_horizon, batch_size, multi_var=True)

# Data windowing on test data
ne_test_ds = window_dataset(ne_test, n_steps, n_horizon, batch_size, multi_var=True)
s_test_ds = window_dataset(s_test, n_steps, n_horizon, batch_size, multi_var=True)
```

## LSTM Model
- NE: MAE on test data = 1191 confirmed cases
- S: MAE on test data = 12752 confirmed cases

### NE region

```python
ne_model = Sequential([
    LSTM(128, activation="relu", return_sequences=True, input_shape= (n_steps, ne_train.shape[1])),
    LSTM(64, return_sequences=False),
    Flatten(),
    Dropout(0.3),
    Dense(128, activation = "relu"),
    Dropout(0.3),
    Dense(128, activation = "relu"),
    Dropout(0.3),
    Dense(64, activation = "relu"),
    Dropout(0.3),
    Dense(32, activation = "relu"),
    Dropout(0.3),
    Dense(n_horizon)
])
```

```python
ne_model.summary()
```

```python
tf.keras.utils.plot_model(ne_model, show_shapes=True)
```

```python
ne_model.compile(
    loss='mean_absolute_error', 
    optimizer=tf.keras.optimizers.Adam(),
    metrics = ['mae']
)
```

```python
early = EarlyStopping("val_loss", patience = 10)
```

```python
ne_model_hist = ne_model.fit(
    ne_train_ds,
    validation_data = ne_val_ds,
    epochs=100,
    callbacks=early
)
```

```python
plt.plot(ne_model_hist.history["loss"])
plt.plot(ne_model_hist.history["val_loss"])
plt.legend(["train", "val"])
plt.show()
```

```python
ne_test_pred = ne_model.predict(ne_test_ds)
```

```python
ne_test_pred = pd.DataFrame(ne_test_pred, columns=['yhat'], index=range(0,len(ne_test_pred)))
```

```python
pd.concat([ne_test.covid_19_confirmed_cases[:41], ne_test_pred["yhat"]], axis=1).plot(figsize=(20, 10))
plt.legend(['test_gt', 'test_pred'])
```

```python
# MAE on test data
mean_absolute_error(ne_test.covid_19_confirmed_cases[:41], ne_test_pred)
```

### S Region

```python
s_model = Sequential([
    LSTM(128, activation="relu", return_sequences=True, input_shape= (n_steps, s_train.shape[1])),
    LSTM(64, return_sequences=False),
    Flatten(),
    Dropout(0.3),
    Dense(128, activation = "relu"),
    Dropout(0.3),
    Dense(128, activation = "relu"),
    Dropout(0.3),
    Dense(64, activation = "relu"),
    Dropout(0.3),
    Dense(32, activation = "relu"),
    Dropout(0.3),
    Dense(16, activation = "relu"),
    Dropout(0.3),
    Dense(8, activation = "relu"),
    Dropout(0.3),
    Dense(n_horizon)
])
```
