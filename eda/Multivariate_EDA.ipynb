# Correlation

## Categorical Variable Correlation
### NE Region
- cat__social_distancing_encounters_grade_A = 0.76
    - Percent change in human encounters compared to pre-COVID-19 period (range from A to F)
        - A: >94% decrease
- cat__social_distancing_encounters_grade_F = -0.76
    - Percent change in human encounters compared to pre-COVID-19 period (range from A to F)
        - F: <40% decrease or increase
- cat__social_distancing_travel_distance_grade_F = -0.78
    - Percent change in average distance traveled compared to pre-COVID-19-period (range from A to F)
        - F: <25% decrease or increase
- cat__social_distancing_total_grade_F = -0.82
    - Average numerical score of the previous three social distancing factors

### S Region
- No highly correlated categorical variables with covid_19_confirmed_cases
