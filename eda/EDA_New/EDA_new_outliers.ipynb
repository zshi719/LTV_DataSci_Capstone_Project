 # Outliers

```python
def outliers(data, column_name, threshold=1.5):
    outliers_df = pd.DataFrame(columns=[column_name, 'date'])
    
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    
    if not outliers.empty:
        outliers_df[column_name] = outliers[column_name]
        outliers_df['date'] = outliers.index
    
    return outliers_df

outlier_ne = outliers(ne, 'covid_19_confirmed_cases', threshold=1.5)
print(outlier_ne)
```

```python
outlier_s = outliers(s, 'covid_19_confirmed_cases', threshold=1.5)
print(outlier_s)
```
